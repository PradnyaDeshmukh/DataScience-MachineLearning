```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
test = [50, 49, 34, 45, 30, 47, 48, 39, 45, 47, 46, 49]
print(test)
```

```python
m = np.mean(test)
v = np.var(test)
print(m,v)
```

```python
getNormalMatrix = np.random.normal(1,1,size=(2,3))
print(getNormalMatrix)
```

```python
sns.distplot(np.random.normal(m,v,size=1000), hist=False)
plt.show()
```

```python
from scipy.stats import skew
skew(test)
```

```python
from scipy.stats import kurtosis
kurtosis(test)
```

```python
data = np.random.normal(m,v, size=(10000))
fig, axs = plt.subplots(figsize = (10,5))
axs.hist(data, bins=90)
```
